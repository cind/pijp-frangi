```python
import nibabel as nib
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
# testing show slice images 
def show_slices(slices):
    # Function to display row of image slices
    #plt.figure(figsize=(10,5))
    fig, axes = plt.subplots(1, len(slices),figsize=(20,10))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

```python
img = nib.load('/Users/nanatang/Documents/GradResearch/ADNI3_samples/sub-4-ad/sub-4-ad-T1rot.nii.gz')
```

```python
data = img.get_fdata()
data_rot = np.rot90(data,axes=(2,1))
np.shape(data)
np.mean(data)
```

```python
# to take center slice, index at half each axis
slice_0 = data[128, :, :]
slice_1 = data[:, 128, :]
slice_2 = data[:, :, 128]
show_slices([slice_0, slice_1, slice_2])
```

```python
# to take center slice, index at half each axis
slice_0 = data_rot[128, :, :]
slice_1 = data_rot[:, 128, :]
slice_2 = data_rot[:, :, 128]
show_slices([slice_0, slice_1, slice_2])
```
